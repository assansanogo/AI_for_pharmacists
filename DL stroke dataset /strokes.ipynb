```python
def var_hist(df, category="smoking_status"):
    variable_hist = df[category].value_counts()
    fig =plt.figure(figsize=(20, 15))
    ax = fig.add_subplot(111)
    print(variable_hist.values)
    print(variable_hist.index)
    ax.bar(variable_hist.index, variable_hist.values, width = 0.35, color=['darkgreen','green','orange'], alpha=0.6)
    plt.title("Repartition variable: " + category, fontsize=29)
    ax.tick_params(labelsize=24)
    plt.show()
```
